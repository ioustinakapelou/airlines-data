```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


dforigin['Delay'] = dforigin['Delay'].apply(lambda x: f"Delay={x}")
dforigin['Origin'] = dforigin['Origin'].apply(lambda x: f"Origin={x}")


transactions = dforigin.values.tolist()

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

df_transformed = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df_transformed, min_support=0.000001, use_colnames=True)

assoc_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7,num_itemsets=10)

rules_sorted = assoc_rules.sort_values(by='confidence', ascending=False)


rules_sorted.iloc[:, 0:7]

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
array = dfpred.values
X = array[:,0:129]
Y = array[:,129]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=7)
dt = DecisionTreeClassifier(random_state=7)


param_grid = {'max_depth': range(1, 10)}  

grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, scoring='accuracy', cv=7)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)

best_dt = grid_search.best_estimator_
test_accuracy = best_dt.score(X_test, y_test)
print("Test Set Accuracy:", test_accuracy)
```

```python
import numpy as np

array = dfencod.values
X = array[:,0:129]
Y = array[:,129]
feature_names = dfencod.columns.tolist()[:129]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=7)
tree = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=7)
tree.fit(X_train, y_train)
importances = tree.feature_importances_
important_features = np.argsort(importances)[::-1]  

N = 5
top_features = [feature_names[i] for i in important_features[:N]]

from sklearn.tree import  export_text
rules = export_text(tree, feature_names=feature_names)

print("\nTop  Most Prominent Rules Including Leaf Nodes")
for line in rules.splitlines():
    if any(feature in line for feature in top_features) or "class:" in line:
        print(line)
```
